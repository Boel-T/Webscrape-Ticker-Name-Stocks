```python
#import dependencies
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Create two empty lists for company names and tickers/symbol
company_name = []
company_ticker = []
```

```python
URL = 'http://www.advfn.com/nyse/newyorkstockexchange.asp?companies=A'
page = requests.get(URL)
soup = BeautifulSoup(page.text, 'html.parser')
odd_rows = soup.find_all('tr', attrs={'class':'ts0'})
even_rows = soup.find_all('tr', attrs={'class':'ts1'})
```

```python
for i in odd_rows:
  row = i.find_all('td')
  company_name.append(row[0].text.strip()) # company name
  company_ticker.append(row[1].text.strip()) # company ticker

  for i in even_rows:
    row = i.find_all('td')
    company_name.append(row[0].text.strip()) # company name
    company_ticker.append(row[1].text.strip()) # company ticker
```

```python
#create a function to scrape the date
def scrape_stock_symbols(Letter):
  Letter = Letter.upper()
  URL = 'http://www.advfn.com/nyse/newyorkstockexchange.asp?companies='+Letter
  page = requests.get(URL)
  soup = BeautifulSoup(page.text, 'html.parser')
  odd_rows = soup.find_all('tr', attrs={'class':'ts0'})
  even_rows = soup.find_all('tr', attrs={'class':'ts1'})

  for i in odd_rows:
    row = i.find_all('td')
    company_name.append(row[0].text.strip()) # company name
    company_ticker.append(row[1].text.strip()) # company ticker

  for i in even_rows:
    row = i.find_all('td')
    company_name.append(row[0].text.strip()) # company name
    company_ticker.append(row[1].text.strip()) # company ticker

  return (company_name, company_ticker)
```

```python
(temp_name, temp_ticker) = scrape_stock_symbols('b')
temp_name
```

```python
#Get a list of every letter in the alphabet
import string
string.ascii_uppercase
```

```python

#loop through all the letters in the alphabet to get all the company names and tickers from the website
for char in string.ascii_uppercase:
  (temp_name, temp_ticker) = scrape_stock_symbols(char)
```

```python
#create a dateFrame that contains the company names and the company tickers
data =pd.DataFrame(columns=['company_name','company_ticker'])
data['company_name'] = temp_name
data['company_ticker'] = temp_ticker
#show data
data
```

```python
#Clean the data
data = data[data['company_name'] != '']
#show data
data
```
